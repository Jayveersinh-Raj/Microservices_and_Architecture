# Challenges
**`Data Managment`** amongst the services is a huge problem for some reasons as mentioned below;
- Bugs in a database (single) can make the entire app crash
- Accessing database of other services is problematic because:
  1. It would crash multiple services together at the same time
  2. A change in the database by Service B team, can trigger bugs for others
